# Exploratory Data Analysis (EDA) – Iris Dataset

**Objective:**  
To explore, visualize, and understand data patterns, feature relationships, and important predictors using Exploratory Data Analysis (EDA).

Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris

%matplotlib inline
```

Load Dataset

```python
iris = load_iris()

df = pd.DataFrame(
    data=iris.data,
    columns=iris.feature_names
)

df['species'] = iris.target
df['species'] = df['species'].map({
    0: 'Setosa',
    1: 'Versicolor',
    2: 'Virginica'
})

df.head()
```

Dataset Information

```python
df.info()
```

Statistical Summary

```python
df.describe()
```

Histogram: Numerical Feature Distribution

```python
df.hist(figsize=(10, 8))
plt.suptitle("Distribution of Numerical Features", fontsize=14)
plt.tight_layout()
plt.show()
```

### Insights
- Petal features show wider spread than sepal features
- Data distribution is close to normal
- Petal dimensions appear more discriminative

Count Plot: Categorical Feature

```python
sns.countplot(x='species', data=df)
plt.title("Species Distribution")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

### Insights
- All three species are equally distributed
- No class imbalance issue


Box Plot: Outlier Detection

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop('species', axis=1))
plt.title("Box Plot for Outlier Detection")
plt.show()
```

### Insights
- Minor outliers present in sepal width
- No extreme anomalies
- Dataset is clean and reliable


Correlation Heatmap

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    df.drop('species', axis=1).corr(),
    annot=True,
    cmap='coolwarm'
)
plt.title("Correlation Heatmap")
plt.show()
```

### Insights
- Strong positive correlation between petal length and petal width
- Sepal features show weak correlation
- Indicates multicollinearity among petal features


Pair Plot

```python
sns.pairplot(df, hue='species')
plt.show()
```

### Important Features for Prediction
Based on EDA:
- Petal Length
- Petal Width

These features show clear class separation and high correlation.

## Final Summary

- Dataset is clean with no missing values
- Balanced class distribution
- Petal features dominate prediction power
- Minimal outliers detected
- Ready for machine learning modeling
